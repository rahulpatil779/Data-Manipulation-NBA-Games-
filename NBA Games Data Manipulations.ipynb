# 8. Reassign `teams` as a subset of its columns 'TEAM_ID', 'CITY', 'NICKNAME'

# 9. Convert both columns `CITY` and `NICKNAME` to a `string` dtype

# 10. Merge (inner) `games` and `teams` based on 'TEAM_ID_home' and 'TEAM_ID', call the merged DataFrame `games_with_home_team`

```python
games_with_home = pd.merge(games,teams,left_on='TEAM_ID_home',right_on='TEAM_ID',how='inner')
```

```python
games_with_home.head()
```

# 11. Rename the column `CITY` as 'city_home', `NICKNAME` as 'nickname_home'

```python
games_with_home=games_with_home.rename(columns={'CITY':'city_home','NICKNAME':'nickname_home'})
```

```python
games_with_home.head()
```

# 12. Merge (inner) `games_with_home_team` and `teams` based on 'TEAM_ID_away' and 'TEAM_ID', call the merged DataFrame `games_with_both_teams`

```python
games_with_both_teams = pd.merge(games_with_home,teams,left_on='TEAM_ID_away',right_on='TEAM_ID')
```

```python
games_with_both_teams.head(2)
```

# 13. Rename the column `CITY` as 'city_away', `NICKNAME` as 'nickname_away'

```python
games_with_both_teams.head(2)
```
